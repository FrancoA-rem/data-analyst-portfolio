```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("netflix_titles.csv")
```

```python
df
```

```python

df['year_added'] = pd.to_datetime(df['date_added'], dayfirst=True, errors='coerce').dt.year.astype('Int64')

estrenos_por_aÃ±o = df['year_added'].value_counts().sort_index()

plt.rcParams['font.family'] = 'DejaVu Sans'

fig, ax = plt.subplots(figsize=(14, 8))

fig.patch.set_facecolor('#0f0f0f')  
ax.set_facecolor('#1c1c1c')         

estrenos_por_aÃ±o.plot(kind='bar', color='#E50914', ax=ax)

ax.set_title("TÃ­tulos aÃ±adidos a Netflix por aÃ±o", fontsize=20, fontweight='bold', color='white')
ax.set_xlabel("AÃ±o", fontsize=14, color='white')
ax.set_ylabel("Cantidad de tÃ­tulos", fontsize=14, color='white')

ax.tick_params(axis='x', colors='white', labelsize=12)
ax.tick_params(axis='y', colors='white', labelsize=12)

plt.xticks(rotation=45)
plt.tight_layout()

plt.savefig("grafico_netflix_linkedin.png", dpi=300, bbox_inches="tight")
plt.show()


```
